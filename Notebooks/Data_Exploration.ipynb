## Test performance of different PDF loaders: PyPDFLoader, Unstructured, Docling

```python
import os
from langchain.document_loaders import PyPDFLoader
loaders = [
    PyPDFLoader("../Data/Budget2024_Documents/fy2024_budget_booklet_english.pdf"),
    PyPDFLoader("../Data/Budget2024_Documents/annexb1.pdf"),
    PyPDFLoader("../Data/Budget2024_Documents/fy2024_budget_statement.pdf")
]
docs = []

for loader in loaders:
    docs.extend(loader.load())

def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)

format_docs(docs)
```

```python
from unstructured.partition.pdf import partition_pdf
import unstructured_pytesseract
#Some dependencies are required to run unstructured library since it leverages on OCR during extraction
unstructured_pytesseract.pytesseract.tesseract_cmd = r'C:/Program Files/Tesseract-OCR/tesseract.exe'
docs = []
loaders = [
    partition_pdf("../Data/Budget2024_Documents/fy2024_budget_booklet_english.pdf", strategy="hi_res"),
    partition_pdf("../Data/Budget2024_Documents/annexb1.pdf", strategy="hi_res"),
    partition_pdf("../Data/Budget2024_Documents/fy2024_budget_statement.pdf", strategy="hi_res")
]

for loader in loaders:
    docs.extend(loader)
# elements = partition_pdf("../Data/Budget2024_Documents/fy2024_budget_booklet_english.pdf", strategy="hi_res")
print("\n\n".join([str(doc) for doc in docs]))
```

```python
from langchain_docling import DoclingLoader
docs = []
loaders = [
    DoclingLoader(file_path = "../Data/Budget2024_Documents/fy2024_budget_booklet_english.pdf"),
    DoclingLoader(file_path = "../Data/Budget2024_Documents/annexb1.pdf"),
    DoclingLoader(file_path = "../Data/Budget2024_Documents/fy2024_budget_statement.pdf")
]

for loader in loaders:
    docs.extend(loader.load())
format_docs(docs)
```
